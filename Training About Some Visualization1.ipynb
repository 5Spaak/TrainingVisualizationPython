# Training of Visualization 


```python
# Importation of libraries
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

## Line Chart

```python
years = range(2000, 2012)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.9372, 0.939]
oranges = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.896, ]
```

```python
sns.set_style("whitegrid")

plt.figure(figsize=(14,6.5))

plt.plot(years, apples, 's-b', linewidth=3, ms=5, mfc='navy')
plt.plot(years, oranges, 'o-r',linewidth=3,ms=9, mew=3)

plt.xlabel("Years")
plt.ylabel("Yield (tons of hectares)")
plt.title('Crop Yields in Kando')

plt.legend(["Apples","Oranges"]);
```

## Scatter plot 

```python
flowers_df = sns.load_dataset("iris")
```

```python
flowers_df
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(13,6))

plt.title("Sepal  Dimensions")
sns.scatterplot(x=flowers_df.sepal_length,
               y=flowers_df.sepal_width,
               hue=flowers_df.species,
               s=100);
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(13,6))

plt.title("Petal  Dimensions")
sns.scatterplot(x=flowers_df.petal_length,
               y=flowers_df.petal_width,
               hue=flowers_df.species,
               s=150);
```

```python
sns.set_style("darkgrid")

plt.figure(figsize=(13,6))
plt.title("Sepal Dimensions")
sns.scatterplot(x="sepal_length",
               y="sepal_width",
               hue="species",
               data=flowers_df,
               s=100);
```
